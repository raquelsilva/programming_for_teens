There is a convention in mathematics to represent the horizontal axis with the letter $x$ and the vertical axis with the letter $y$. So, now, let's create a two dimensional graph using random values created by the computer for the axis $x$ and $y$.

```python
import random

x = random.sample(range(1, 10), 9)
y = random.sample(range(1, 10), 9)
```

Let's check the numbers created above:

```python
print x
```

```python
print y
```

Now, let's see how this numbers appear in a graph of 2 dimensions.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

font1 = {'family': 'serif',                # set the color of the text
        'color':  'green',
        'weight': 'normal',
        'size': 16,
        }

font2 = {'family': 'serif',                # set the color of the text
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }

ax.set_ylim(0, 10)                      # set the maximum and minimum values for the y axis
ax.set_xlim(0, 10)                      # set the maximum and minimum values for the x axis

plt.xlabel('X Axis', fontdict=font1)       # Text to identify the axis
plt.ylabel('Y Axis', fontdict=font2)       # Text to identify the axis

ax.grid(True)                             # Add internal lines in the graph

ax.axhline(linewidth=4, color="g")        # inc. width of x-axis and color it green
ax.axvline(linewidth=4, color="b")        # inc. width of y-axis and color it red

plt.plot(x, y, 'ro')                      # plots the values of x and y and use the marker in red color

plt.show()
```

Now I will create a graph in three dimensions using the same data above but adding one more axis that it is called $z$. So, I will create now a list of values for $z$, also using random numbers.

```python
z = random.sample(range(1, 10), 9)
```

```python
print z
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

font1 = {'family': 'serif',                # set the color of the text
        'color':  'green',
        'weight': 'normal',
        'size': 16,
        }

font2 = {'family': 'serif',                # set the color of the text
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }

font3 = {'family': 'serif',                # set the color of the text
        'color':  'magenta',
        'weight': 'normal',
        'size': 16,
        }

ax.set_ylim(0, 10)                      # set the maximum and minimum values for the y axis
ax.set_xlim(0, 10)                      # set the maximum and minimum values for the x axis
ax.set_zlim(0, 10)                      # set the maximum and minimum values for the z axis

plt.xlabel('X Axis', fontdict=font1)       # Text to identify the axis
plt.ylabel('Y Axis', fontdict=font2)       # Text to identify the axis
plt.ylabel('Z Axis', fontdict=font3)       # Text to identify the axis

ax.grid(True)                             # Add internal lines in the graph

plt.plot(x, y, z, 'ro')                      # plots the values of x and y and use the marker in red color

plt.show()
```

But integers are a set of numbers - represented by the symbol $\mathbb{Z}$ - which include both natural numbers (positive) and negative numbers. So, let's now create a graph in two dimensions to include both positive and negative numbers.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

font1 = {'family': 'serif',                # set the color of the text
        'color':  'green',
        'weight': 'normal',
        'size': 16,
        }

font2 = {'family': 'serif',                # set the color of the text
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }

ax.set_ylim(-10, 10)                      # set the maximum and minimum values for the y axis
ax.set_xlim(-10, 10)                      # set the maximum and minimum values for the x axis

plt.xlabel('X Axis', fontdict=font1)       # Text to identify the axis
plt.ylabel('Y Axis', fontdict=font2)       # Text to identify the axis

ax.grid(True)                             # Add internal lines in the graph

ax.axhline(linewidth=4, color="g")        # inc. width of x-axis and color it green
ax.axvline(linewidth=4, color="b")        # inc. width of y-axis and color it red

plt.plot(x, y, 'ro')                      # plots the values of x and y and use the marker in red color

plt.show()
```

Note that all the dots are plotted at one of the four quadrants of the graph. It happens when $x$ is positive and $y$ is positive.

```python
x_n = random.sample(range(-10, -1), 9)
y_n = random.sample(range(-10, -1), 9)
```

Where it could be applied? I have an idea. Let'a use the planet Earth as our expanded graph. The $y$ axis will be our Greenwich meridian, also called "Prime Meridian", and the $x$ axis will be the Equator line. See the image bellow to visualize the axis and compare with what I did above:

```python
Image("images/prime_meridian_and_equator.jpg")
```

The Equator divides the world in North and South. The Prime Meridian divides the world in East and West.

The world time is based on the Prime Meridian, which is considered reference for zero in relation with East and West sides. See the image bellow with the world time zones.

```python
Image("images/fuseaux.jpg")
```

So, the numbers which vary from -12 to 12, form the $y$ axis of the world. But, as you can see by the colors, the time zones are not exactly straight. This is due to the countries limits or even, provinces, if the country is too big as Brazil, USA, Canada, Russia, Austrália, etc. For geographic purposes other reference is used though. It is called longitude. In the image above the longitude values can be seen varying from 0 to 180. But this is one kind of notation. You can also use negative numbers to represent the western world and positive numbers to represent the eastern world. If you only use positive numbers it is necessary to include the quadrants: N (North), S (South), W (West) and/or E (East). For example, what is the latitude of Iceland? Well, it could be $15^\circ$W or only $-15^\circ$. For mathematical operations the negative numbers are more useful, though.

The notation for the Equator is very similar, but instead of longitude we say latitude. See the image below:

```python
Image("images/imagee.jpg")
```

Note that here the negative numbers are not being used. The Equator is represented by zero and other two other lines are considered important: the Tropic of Cancer and the Tropic of Capricorn, located at $23,5^\circ$N and $23,5^\circ$S, respectively. The region between the two tropics is known as tropical region. Above and below the tropics we find the temperate regions and, near the poles North and South we find the polar regions. Mathematically, the North region (or hemisphere) is represented by positive numbers and the South hemisphere by negative numbers.

We can also use the surface of the planet as a reference for positive and negative numbers. For example, we represent the altitude above the surface as positive and below the surface as negative.
